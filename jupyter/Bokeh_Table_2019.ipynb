```python
import bokeh.plotting as plt
import bokeh.io as bio
import bokeh.models as bmo
import bokeh.models.widgets as bmw
import ipywidgets as widgets
import pandas as pd

import server.model.event as sme
import server.view.dataframes as svd

bio.output_notebook()
```

```python
evt = sme.EventDal.get_current_event()
mtch = sme.EventDal.get_current_match()
txt_event = widgets.Text(description='Event Code:', value=evt[1])
txt_season = widgets.IntText(description='Season:', value=int(evt[2]))
txt_match = widgets.Text(description='Match:', value=mtch)
btn_event = widgets.Button(description='Select Event', value=mtch)

def btn_event_click(btn):
    sme.EventDal.set_current_event(txt_event.value, txt_season.value)
    sme.EventDal.set_current_match(txt_match.value)
    print('Selected Event:', sme.EventDal.get_current_event(),
          'Current Match:', sme.EventDal.get_current_match())
    
btn_event.on_click(btn_event_click)

widgets.HBox([txt_event, txt_season, txt_match, btn_event])
```

```python
rank_df = svd.ranking_df(12).fillna(0)
rank_df.head()
```

```python
rank_df.xs('climb', axis=1, level=2)
```

```python
rank_cds = bmo.ColumnDataSource(rank_df)
rank_cds.column_names
```

```python
cols = ['team', 'auto_robot_centerCargo_avg_successes']
fixed2 = bmw.NumberFormatter(format='0.00')
cols = [
        bmw.TableColumn(field='team', title='Team'),
        bmw.TableColumn(field='auto_robot_centerCargo_avg_successes', title='Center Cargo Auto', formatter=fixed2),
        bmw.TableColumn(field='auto_robot_centerHatch_avg_successes', title='Center Hatch Auto', formatter=fixed2),
        bmw.TableColumn(field='auto_robot_closeCargo_avg_successes', title='Close Cargo Auto', formatter=fixed2),
        bmw.TableColumn(field='teleop_robot_rocketHatch3_avg_successes', title='Rocket H3', formatter=fixed2),
        bmw.TableColumn(field='finish_robot_climb_avg_capabilities', title='Climb', formatter=fixed2),
    ]
rank_tbl = bmw.DataTable(source=rank_cds, columns=cols)
bio.output_notebook()
bio.show(rank_tbl)
```
