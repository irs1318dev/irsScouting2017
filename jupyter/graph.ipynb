```python
import pandas as pd
import psycopg2
import bokeh.plotting
import bokeh.models as bmodels
import bokeh.plotting as plt
import bokeh.palettes as bpalettes
import bokeh.transform as btransform
import bokeh.io

import server.model.connection as smc
import server.model.version as smv
import server.model.event as sme

bokeh.io.output_notebook()
```

```python
conn = smc.pool.getconn()
```

```python
sme.EventDal.set_current_event("test_data","2019")
sme.EventDal.set_current_match("020-q")
```

```python
sql = '''
    SELECT * FROM vw_measures 
    WHERE task IN ('centerHatch', 'farHatch', 'closeHatch', 'rocketHatch',
    'rocketHatch1', 'rocketHatch2', 'rocketHatch3', 'csHatch');
'''
pd.read_sql(sql, conn)
```

```python
match = '020-q'
alliance = 'red'
sql = '''
    SELECT team, alliance FROM schedules
        INNER JOIN status ON schedules.event_id=status.event_id 
    WHERE schedules.event_id=status.event_id AND schedules.match=%s AND alliance=%s;

'''
pd.read_sql(sql, conn, params=[match, alliance])
```

```python
match = '020-q'
alliance = 'red'
sql = '''
    SELECT team, task, SUM(successes) AS hatches_placed FROM vw_measures 
    WHERE task IN ('centerHatch', 'farHatch', 'closeHatch', 'rocketHatch',
    'rocketHatch1', 'rocketHatch2', 'rocketHatch3', 'csHatch')
    AND team IN (SELECT team FROM schedules
                 INNER JOIN status ON schedules.event_id=status.event_id 
                 WHERE schedules.event_id=status.event_id AND schedules.match=%s
                 AND alliance=%s)
    GROUP BY team, task;
'''
pd.read_sql(sql, conn, params=[match, alliance])
```

```python
match = '020-q'
alliance = 'red'
sql = '''
    SELECT team, SUM(successes) AS hatches_placed, SUM(attempts) AS attempts FROM vw_measures 
    WHERE task IN ('centerHatch', 'farHatch', 'closeHatch', 'rocketHatch',
    'rocketHatch1', 'rocketHatch2', 'rocketHatch3', 'csHatch')
    AND team IN (SELECT team FROM schedules
                 INNER JOIN status ON schedules.event_id=status.event_id 
                 WHERE schedules.event_id=status.event_id AND schedules.match=%s
                 AND alliance=%s)
    GROUP BY team;
'''
df_temp = pd.read_sql(sql, conn, params=[match, alliance])
df_redHatches = df_temp.set_index('team')
df_redHatches
```

```python
match = '020-q'
alliance = 'blue'
sql = '''
    SELECT team, SUM(successes) AS hatches_placed, SUM(attempts) AS attempts FROM vw_measures 
    WHERE task IN ('centerHatch', 'farHatch', 'closeHatch', 'rocketHatch',
    'rocketHatch1', 'rocketHatch2', 'rocketHatch3', 'csHatch')
    AND team IN (SELECT team FROM schedules
                 INNER JOIN status ON schedules.event_id=status.event_id 
                 WHERE schedules.event_id=status.event_id AND schedules.match=%s
                 AND alliance=%s)
    GROUP BY team;
'''
df_temp = pd.read_sql(sql, conn, params=[match, alliance])
df_blueHatches = df_temp.set_index('team')
df_blueHatches
```

```python
df_rHatches = df_redHatches.T
df_rHatches
```

```python
df_rHatches.index.name
```

```python
rHatches_cds = bmodels.ColumnDataSource(df_rHatches)
rHatches_cds.column_names
```

```python
rteams = rHatches_cds.column_names[1:]
rteams
```

```python
p = plt.figure(title='Hatches Placed', x_range=rHatches_cds.data['index'], 
              plot_width=700, plot_height=350)
p.vbar_stack(rteams, x='index', width=0.8, 
            source=rHatches_cds, color=bpalettes.Reds3, legend=[" " + x for x in rteams])

bokeh.io.show(p)
```

```python
df_bHatches = df_blueHatches.T
df_bHatches
```

```python
bHatches_cds = bmodels.ColumnDataSource(df_bHatches)
bHatches_cds.column_names
```

```python
bteams = bHatches_cds.column_names[1:]
bteams
```

```python
p = plt.figure(title='Hatches Placed', x_range=bHatches_cds.data['index'], 
              plot_width=700, plot_height=350)
p.vbar_stack(bteams, x='index', width=0.8, 
            source=bHatches_cds, color=bpalettes.Blues3, legend=[" " + x for x in bteams])

bokeh.io.show(p)
```

```python
p = plt.figure(title='Hatches Placed', x_range=['hatches_placed', 'attempts'], 
              plot_width=700, plot_height=350)
p.vbar_stack(bteams, x=btransform.dodge('index',  0.25,  range=p.x_range), width=0.4, 
            source=bHatches_cds, color=bpalettes.Blues3, legend=[" " + x for x in bteams])
p.vbar_stack(rteams, x=btransform.dodge('index',  -0.25,  range=p.x_range), width=0.4, 
            source=rHatches_cds, color=bpalettes.Reds3, legend=[" " + x for x in rteams])

bokeh.io.show(p)
```
